```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d.axes3d import Axes3D
from os import listdir
from IPython.display import HTML

%matplotlib notebook
```

```python
# The Tx-Rx distances at which there's a local maxima/minima
# [min_0, max_1 ..., min_12]
distances = np.array([2.5, 4.5, 7.0, 9.0, 11.0, 13.5, 15.5, 18.0, 20.0, 22.0, 25.0, 27.0, 29.0])
```

```python
# Data Array: Rx-Voltage
    # 0 -> V_Rx @ d=2.5mm, 1 -> V_Rx @ d=4.5mm, ... 12 -> 0 -> V_Rx @ d=29.0mm

data_arr = np.zeros([len(distances),32*5004])

i = 0
for distance in distances:
    distance = str(distance).replace('.', '-')
    measurements = np.array(listdir("../measurements/csv/" + 
                                    distance + "/"))
    j = 0
    for measurement in measurements:
        csv_data = pd.read_csv("../measurements/csv/" + 
                               distance + "/" +
                               measurement, 
                               names=["Time", "Rx Voltage (mV)"])
        csv_data = csv_data[2:].drop(["Time"], axis=1).reset_index(drop=True)
        csv_data = csv_data.astype("float64").to_numpy().flatten()
        
        if '4-5' in measurement:
            csv_data = csv_data*1000
        
        data_arr[i][j*5004:(j+1)*5004] = csv_data
        j += 1
    i += 1
```

```python
rx_voltages = np.empty(len(distances))

for i in range(len(distances)):
    rx_voltage = (data_arr[i].max()-data_arr[i].min())/2
    rx_voltages[i] = rx_voltage
```
